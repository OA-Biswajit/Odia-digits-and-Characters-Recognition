```python
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
```

```python
datagen = ImageDataGenerator(
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        rescale=1./255,
        shear_range=0.1,
        zoom_range=0.1,
        horizontal_flip=False,
        fill_mode='nearest')
```

```python
for i in range(57):
    for j in range(5):
        pic = load_img(f'data/{i}/{j}.jpg')
        pic_array = img_to_array(pic)
        pic_array =pic_array.reshape((1,) + pic_array.shape) # Converting into 4 dimension array
        count = 0
        for batch in datagen.flow(pic_array, batch_size=5,save_to_dir=f"new_data/{i}", save_prefix=f"gen_{j}", save_format='jpg'):
          count += 1
          if count >= 10:
              break
```
