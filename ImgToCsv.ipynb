```python
data_dir = 'new_data'
image_size = (100, 100)
num_folders = 57
num_images_per_folder = 55
# Initialize an empty array to store the images
image_array = np.empty((num_folders, num_images_per_folder, *image_size,3), dtype=np.uint8)

# Loop through each folder
for folder_index in range(num_folders):
    folder_path = os.path.join(data_dir, str(folder_index))
    image_index=0
    # Loop through each image in the folder
    for filename in os.listdir(folder_path):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(folder_path, filename)
        # Read the image using OpenCV
            image = cv2.imread(image_path)
        
        # Resize the image to the desired size
            image = cv2.resize(image, image_size)

        # Store the image in the array
            image_array[folder_index, image_index] = image
            image_index+=1
# Print the shape of the image array
print(image_array.shape)
```
